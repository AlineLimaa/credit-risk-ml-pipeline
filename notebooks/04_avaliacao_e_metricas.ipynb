# Avaliação de Modelos e Métricas

## Objetivo
A avaliação correta dos modelos é essencial em problemas de risco de crédito, onde erros possuem custos diferentes.

Neste notebook, analisamos:
- Métricas como Accuracy, Precision, Recall, F1-Score e ROC-AUC
- Matriz de confusão
- Comparação entre modelos

O foco é entender não apenas qual modelo performa melhor, mas **por quê**.
